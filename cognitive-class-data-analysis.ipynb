# Module 2
## Pre-processing Data in Python
### Dealing with Missing Values

```python
# drop missing values (axis = 0 drop entire row, axis = 1 drop entire column)
df.dropna(subset=["price"], axis = 0, inplace = True);
```

```python
import numpy as np

# convert "?" to NaN
df.replace("?", np.nan, inplace = True)
df.head()
```

```python
# evaluating for missing data ("True" stands for missing value, while "False" stands for not missing value)
missing_data = df.isnull()
missing_data.head()
```

```python
# count missing values in each column
for column in missing_data.columns.values.tolist():
    print(column)
    print(missing_data[column].value_counts())
    print("")
```

```python
# count "NaN" values before the replace
print("check 'NaN' values:", df["normalized-losses"].isna().sum())

# calculate the average of the column
avg_1 = df["normalized-losses"].astype("float").mean(axis = 0)
print("avg_1 value: ", avg_1)

# replace "NaN" by mean value in "normalized-losses" column
df["normalized-losses"].replace(np.nan, avg_1, inplace = True)

# recheck "NaN" values after the replace
print("recheck 'NaN' values:", df["normalized-losses"].isna().sum())
```

```python
# check which values are present in a particular column
df['num-of-doors'].value_counts()
```

```python
# check the most common type automatically
df['num-of-doors'].value_counts().idxmax()
```

```python
# check "NaN" values before the replace
print("check 'NaN' values:", df["num-of-doors"].isna().sum())

#replace the missing 'num-of-doors' values by the most frequent 
df["num-of-doors"].replace(np.nan, "four", inplace = True)

# recheck which values are present in a particular column
print(df['num-of-doors'].value_counts())
```

```python
# check "NaN" values before the replace
print("check 'NaN' values:", df["price"].isna().sum())

# simply drop whole row with NaN in "price" column
df.dropna(subset=["price"], axis=0, inplace = True)

# reset index, because we droped two rows
df.reset_index(drop = True, inplace = True)

# recheck "NaN" values after the replace
print("recheck 'NaN' values:", df["price"].isna().sum())
```

### Dealing with Data Formatting

```python
# checking data format
df.dtypes
```

```python
# correct data format
df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
df[["price"]] = df[["price"]].astype("float")
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")

print("Done!")
```

```python
# recheck data types
df.dtypes
```

### Dealing with Data Normalization

```python
df.head()
```

```python
# transform mpg to L/100km
df['city-L/100km'] = 235/df["city-mpg"]

# check your transformed data
df.head()
```

```python
# replace (origianl value) by (original value)/(maximum value)
df['length'] = df['length']/df['length'].max()
df['height'] = df['height']/df['height'].max()
df['width'] = df['width']/df['width'].max()
df[["length","width","height"]].head()
```

### Dealing with Data Binning

```python
# convert data to correct format 
df["horsepower"]=df["horsepower"].astype(float, copy=True)
binwidth = (max(df["horsepower"])-min(df["horsepower"]))/4
bins = np.arange(min(df["horsepower"]), max(df["horsepower"]), binwidth)
bins
```

```python
# set group names
group_names = ['Low', 'Medium', 'High']

df['horsepower-binned'] = pd.cut(df['horsepower'], bins, labels=group_names,include_lowest=True )
df[['horsepower','horsepower-binned']].head(20)
```
